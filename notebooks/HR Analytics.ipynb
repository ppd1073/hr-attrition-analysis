```python
# Step 1: Impute missing performance fields
rating_cols = [
    'EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction',
    'TrainingOpportunitiesWithinYear', 'TrainingOpportunitiesTaken',
    'WorkLifeBalance', 'ManagerRating', 'SelfRating'
]

for col in rating_cols:
    df7[col] = df7[col].fillna(df7[col].median())

# Step 2: Create flag for missing performance review
df7['MissingPerformanceReview'] = df7['ReviewDate'].isna().astype(int)

# Step 3: Drop review metadata
df7 = df7.drop(columns=['ReviewDate', 'PerformanceID'])

# Final Check
print("Remaining missing values:", df7.isnull().sum().sum())  # Should be 0
print("Final shape:", df7.shape)
```

```python
df7['AttritionFlag'] = df7['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# List of categorical columns to encode
categorical_cols = [
    'Gender', 'Department', 'JobRole', 'EducationField', 'MaritalStatus',
    'BusinessTravel', 'OverTime','State','Ethnicity'
]

df_model = pd.get_dummies(df7, columns=categorical_cols, drop_first=True)

```

```python
drop_cols = ['EmployeeID', 'FirstName', 'LastName', 'Attrition', 'AgeGroup', 'SalaryBand','HireDate']
df_model = df_model.drop(columns=drop_cols, errors='ignore')
```

```python
X = df_model.drop(columns='AttritionFlag')
y = df_model['AttritionFlag']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.3, random_state=42)
```

```python
non_numeric_cols = X.select_dtypes(include='object').columns
print("Non-numeric columns:", non_numeric_cols.tolist())
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
rf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

y_pred = rf.predict(X_test)
y_proba = rf.predict_proba(X_test)[:, 1]

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=3))
print("\nROC AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Get feature importances from trained model
feat_importance = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)

# Plot Top 15 Features
plt.figure(figsize=(10, 6))
sns.barplot(x=feat_importance.values[:15], y=feat_importance.index[:15], palette="coolwarm")
plt.title("Top 15 Features Influencing Attrition")
plt.xlabel("Feature Importance")
plt.tight_layout()
plt.show()
```

```python
# Get prediction probabilities (risk of attrition)
df7['AttritionRiskScore'] = rf.predict_proba(X)[:, 1]  # Probability of class 1 = leaving

# Define a threshold for "high risk"
risk_threshold = 0.70
df7['HighAttritionRisk'] = (df7['AttritionRiskScore'] > risk_threshold).astype(int)

# Select fields HR will care about
risk_table = df7[['EmployeeID', 'FirstName', 'LastName', 'JobRole', 'Department',
                  'Age', 'Gender', 'AttritionRiskScore', 'HighAttritionRisk']]

# Sort by highest risk
risk_table = risk_table.sort_values(by='AttritionRiskScore', ascending=False)

# Show top 10 highest risk employees
risk_table.head(10)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score

# One-hot encode again on deduplicated df7
df_model = pd.get_dummies(df7, columns=[
    'Gender', 'Department', 'JobRole', 'EducationField',
    'MaritalStatus', 'BusinessTravel', 'OverTime', 'State', 'Ethnicity'
], drop_first=True)

# Drop ID and non-useful columns
df_model = df_model.drop(columns=[
    'EmployeeID', 'FirstName', 'LastName', 'Attrition',
    'SalaryBand', 'AgeGroup', 'HireDate'
], errors='ignore')

# Target & features
X = df_model.drop(columns='AttritionFlag')
y = df_model['AttritionFlag']

leakage_cols = ['AttritionRiskScore', 'HighAttritionRisk']
df_model = df_model.drop(columns=leakage_cols, errors='ignore')

X = df_model.drop(columns='AttritionFlag')
y = df_model['AttritionFlag']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.3, random_state=42)

# Re-train
rf = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
rf.fit(X_train, y_train)

##

y_pred = rf.predict(X_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=3))


y_proba = rf.predict_proba(X_test)[:, 1]
print("ROC AUC Score:", roc_auc_score(y_test, y_proba))

```

```python
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight={0: 1, 1: 3}  # penalize missed leavers more
)

rf.fit(X_train, y_train)  # retrain using the new class weights

# Then evaluate
y_pred = rf.predict(X_test)

from sklearn.metrics import confusion_matrix, classification_report

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, digits=3))

```

```python
from sklearn.preprocessing import StandardScaler
df7['OverTimeFlag'] = df7['OverTime'].map({'Yes': 1, 'No': 0})

# Select meaningful numeric features
cluster_features = df7[[
    'AttritionRiskScore',
    'YearsAtCompany',
    'YearsSinceLastPromotion',
    'Salary',
    'JobSatisfaction',
    'WorkLifeBalance',
    'EnvironmentSatisfaction',
    'OverTimeFlag',
    'ManagerRating'
]].copy()

# Fill any NaNs (if any)
cluster_features.fillna(cluster_features.median(), inplace=True)

# Scale the data
scaler = StandardScaler()
X_cluster = scaler.fit_transform(cluster_features)
```

```python
from sklearn.cluster import KMeans

# Choose number of clusters (tune if needed)
kmeans = KMeans(n_clusters=3, random_state=42)
df7['Cluster'] = kmeans.fit_predict(X_cluster)
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Reduce to 2 dimensions for plotting
pca = PCA(n_components=2)
components = pca.fit_transform(X_cluster)

df7['PC1'] = components[:, 0]
df7['PC2'] = components[:, 1]

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df7, x='PC1', y='PC2', hue='Cluster', palette='coolwarm', size='AttritionRiskScore', sizes=(20, 200), alpha=0.8)
plt.title('High-Risk Employee Clusters (PCA-Reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```
